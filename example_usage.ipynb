# Using the Ook Translator

This notebook shows how to import and run the translation functions from the Ook Translator.
We will translate between human text and Orangutan.


```python
# Import the translation functions
from src.ook_translator import translate_human, translate_orangutan
```

## Translate from human to Orangutan
We will use the example file `example_message.txt` provided in the data folder.   
This file contains the text "Hello world".  
The expected output is "OokkoKoOKkoOKkok ookkokoOkKoOKkOk" 

```python
# Translate the orangutan text to human
translate_human.translate_human("data/example.txt", "data/example_translated.txt")

# Show contents of the created file
!cat data/example_translated.txt
```

## Translate from orangutan to human
We will use the example file `ok.txt` provided in the data folder.  
This file contains the text "Hello world".  
The expected output is "<HELLA, HELLO, HILLA, HILLO> <WORLD, WURLD>" from which we can read "Hello world"

```python
# Translate the orangutan text to human
translate_orangutan.translate_orangutan("data/ok.txt", "data/ok_translated.txt")

# Show contents of the created file
!cat data/ok_translated.txt
```
